```python
import os
import pandas as pd
import requests
# ---------------- CONFIG ----------------
INPUT_FILE = "train(1).xlsx"   # dataset file
OUTPUT_DIR = "images"       # folder to save images
IMAGE_SIZE = 400            # image resolution
DELTA = 0.002               # zoom control
# ----------------------------------------

# Create images folder if it doesn't exist
os.makedirs(OUTPUT_DIR, exist_ok=True)

# Load dataset
df = pd.read_excel(INPUT_FILE)

def fetch_satellite_image(lat, lon, idx):
    """
    Fetch satellite image for given latitude & longitude
    using ESRI World Imagery (no API key required)
    """

    # Define bounding box around the coordinate
    bbox = f"{lon-DELTA},{lat-DELTA},{lon+DELTA},{lat+DELTA}"

    # Satellite image URL
    url = (
        "https://services.arcgisonline.com/ArcGIS/rest/services/"
        "World_Imagery/MapServer/export"
        f"?bbox={bbox}"
        "&bboxSR=4326"
        "&imageSR=4326"
        f"&size={IMAGE_SIZE},{IMAGE_SIZE}"
        "&format=png"
        "&f=image"
    )

    # Download image
    response = requests.get(url, timeout=10)

    # Save image if successful
    if response.status_code == 200:
        with open(f"{OUTPUT_DIR}/house_{idx}.png", "wb") as f:
            f.write(response.content)
    else:
        print(f"Failed to fetch image for index {idx}")

# Loop over all rows in dataset
for idx, row in df.head(300).iterrows():
    fetch_satellite_image(row["lat"], row["long"], idx)

print("Satellite image download completed.")
```
